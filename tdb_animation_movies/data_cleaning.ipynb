```python
import pandas as pd
import numpy as np
import json
```

```python
with open('films_animation_complet_détails.json', 'r') as file:
    data = json.load(file)
```

```python
df = pd.json_normalize(data)
df.columns
```

```python
df.drop(columns=['adult','backdrop_path','homepage','imdb_id','overview','popularity','poster_path','video','belongs_to_collection.id',
                 'belongs_to_collection.name','belongs_to_collection.poster_path','belongs_to_collection.backdrop_path','belongs_to_collection','id',
                'spoken_languages','status','origin_country']
          ,inplace = True)
```

```python
df.head()
```

```python
def explode (data,nom_colonne) :

    # "Exploser" la colonne (au cas où elle contiendrait des listes)
    data_exploded = data.explode(nom_colonne).reset_index(drop=True)
    
    # Aplatir la colonne imbriquée
    tempo_df = pd.json_normalize(data_exploded[nom_colonne])

    # Supprimer la colonne originale imbriquée et ajouter les nouvelles colonnes aplaties
    data_exploded = data_exploded.drop(columns=[nom_colonne])
    data = pd.concat([data_exploded, tempo_df], axis=1)

    data[f"id_{nom_colonne}"] = data['id']
    data[f"name_{nom_colonne}"] = data['name']

    data.drop(columns=['id','name'], inplace = True)

    
    return data
```

```python
def explode_bis (data,nom_colonne) :

    # "Exploser" la colonne (au cas où elle contiendrait des listes)
    data_exploded = data.explode(nom_colonne).reset_index(drop=True)
    
    # Aplatir la colonne imbriquée
    tempo_df = pd.json_normalize(data_exploded[nom_colonne])

    # Supprimer la colonne originale imbriquée et ajouter les nouvelles colonnes aplaties
    data_exploded = data_exploded.drop(columns=[nom_colonne])
    data = pd.concat([data_exploded, tempo_df], axis=1)

    data[f"name_{nom_colonne}"] = data['name']
    data[f"id_{nom_colonne}"] = data['id']
    data[f"origin_country_{nom_colonne}"] = data['origin_country']
    
    data.drop(columns=['name','origin_country'], inplace = True)

    
    return data
```

```python
def explode_tiers (data,nom_colonne) :

    # "Exploser" la colonne (au cas où elle contiendrait des listes)
    data_exploded = data.explode(nom_colonne).reset_index(drop=True)
    
    # Aplatir la colonne imbriquée
    tempo_df = pd.json_normalize(data_exploded[nom_colonne])

    # Supprimer la colonne originale imbriquée et ajouter les nouvelles colonnes aplaties
    data_exploded = data_exploded.drop(columns=[nom_colonne])
    data = pd.concat([data_exploded, tempo_df], axis=1)

    data[f"name_{nom_colonne}"] = data['name']

    data.drop(columns=['name'], inplace = True)

    
    return data
```

```python
df_trans = explode(df,'genres')
df_trans = explode_bis(df_trans,'production_companies')
df_trans = explode_tiers(df_trans,'production_countries')

df_trans.head(3)
```

```python
#Supprimer les colonnes indésirables
df_trans.drop(columns=['logo_path','id','iso_3166_1'], inplace = True)
```
